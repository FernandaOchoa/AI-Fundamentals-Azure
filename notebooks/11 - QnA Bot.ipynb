# IA de conversación

Piense en la frecuencia con la que se comunica con otras personas mediante servicios de mensajería instantánea, redes sociales, correo electrónico u otras tecnologías en línea. Para muchos, es la forma de contacto más habitual. Cuando uno tiene una duda en el trabajo, puede preguntar a un compañero a través de un mensaje de chat, que se puede enviar incluso desde dispositivos móviles, por lo que es posible estar siempre en contacto.

Los bots son agentes con inteligencia artificial capaces de comunicarse a través de estos canales, lo que permite tener una conversación natural con los servicios de software.

## Crear una base de conocimiento de QnA Maker

Para casos de asistencia al cliente, es muy habitual crear un bot que pueda interpretar preguntas frecuentes y responder a ellas a través de un chat en el sitio web, por correo electrónico o mediante una interfaz de voz. Detrás de la interfaz del bot hay una base de conocimiento de preguntas y las respuestas adecuadas entre las que el bot puede buscar.

El servicio QnA Maker forma parte de Azure Cognitive Services y permite crear rápidamente una base de conocimiento, ya sea mediante la introducción de pares de preguntas-respuesta o mediante un documento o sitio web existentes. Después, puede utilizar sus funciones integradas de procesamiento del lenguaje natural para interpretar las preguntas y encontrar las respuestas adecuadas.

1. Abra otra pestaña del navegador y vaya al portal de QnA Maker (https://qnamaker.ai). Inicie sesión con la cuenta de Microsoft asociada a su suscripción de Azure.
2. En el portal de QnA Maker, seleccione **Create a knowledge base**.
3. Si aún no ha creado un recurso del servicio de QnA Maker, seleccione **Create a QnA service**. Azure Portal se abrirá en otra pestaña para que pueda crear un servicio de QnA Maker en su suscripción. Use la configuración siguiente:
   - **Administrado (versión preliminar)**: sin seleccionar. 
   - **Suscripción**: *su suscripción de Azure*
   - **Grupo de recursos**: *seleccione un grupo de recursos existente o cree uno nuevo*
   - **Nombre**: *un nombre único para su recurso de QnA*
   - **Plan de tarifa**: F0
   - **Plan de tarifa de Azure Search**: F
   - **Ubicación de Azure Search**: *cualquier ubicación disponible*
   - **Nombre de la aplicación**: _el mismo que el utilizado en **Name** (".azurewebsites.net" se utilizará automáticamente)_
   - **Ubicación del sitio web**: _la misma que la utilizada en **Azure Search location**_
   - **Datos de la aplicación**: deshabilitar

> **Nota**: Si ya ha aprovisionado recursos gratuitos de **QnA Maker** o **Azure Search**, es posible que su cuota no permita crear más. En ese caso, deberá seleccionar un plan distinto a **F0**/**F**.

4. Espere a que se complete la implementación del servicio de QnA y de los recursos relacionados en Azure Portal.
5. Vuelva a la pestaña del portal de QnA Maker y, en la sección **Step 2**, haga clic en **Refresh** para actualizar la lista de recursos del servicio de QnA disponibles.
6. Conecte su servicio de QnA a su base de conocimientos. Para ello, seleccione estas opciones:
  - **Microsoft Azure Directory ID**: *el identificador del directorio de Azure de su suscripción*
  - **Azure subscription name**: *su suscripción de Azure*
  - **Azure QnA service**: *el recurso del servicio de QnA creado en el paso anterior*
  - **Language**: English

### (!) Comprobar
Si aparece un mensaje que indica que el rol no tiene los permisos necesarios para realizar la acción actual, actualice la página del navegador del portal de QnA Maker.

7. En **Step 3**, escriba el nombre **Margie's Travel KB**.
8. En **Step 4**, en el cuadro para la **URL**, escriba <a href=''>*https://github.com/MicrosoftDocs/ai-fundamentals/raw/master/data/qna_bot/margies_faq.docx*</a> y haga clic en **+ Add URL**. Después, en **Chit-chat**, seleccione **Professional**.
9. En **Step 5**, haga clic en **Create your KB**.
10. Espere alrededor de un minuto mientras se crea su base de conocimiento. Después, revise las preguntas y respuestas importadas desde el documento de preguntas frecuentes y las respuestas de chat profesionales predefinidas.

## Editar la base de conocimiento

Su base de conocimiento se alimenta de los detalles del documento de preguntas frecuentes y de algunas respuestas predefinidas. Puede agregar pares de pregunta-respuesta para complementar su base de conocimiento.

1. Haga clic en **+ Add QnA Pair**.
2. En el cuadro **Question**, escriba `Hello`. Después, haga clic en **+ Add alternative phrasing** y escriba `Hi`.
3. En el cuadro **Answer**, escriba `Hello`.

## Entrenar y probar la base de conocimiento

Ahora que ha creado una base de conocimiento, es el momento de probarla en el portal de QnA Maker.

1. En la esquina superior derecha de la página, haga clic en **Save and train** para entrenar su base de conocimiento. Para ver el botón, es posible que tenga que maximizar la ventana.
2. Después de completar el entrenamiento, haga clic en **&larr; Test** para abrir el panel de pruebas.
3. En la parte inferior del panel de pruebas, escriba el mensaje *Hi*. Después de hacerlo, debería aparecer la respuesta **Hello**.
4. En la parte inferior del panel de pruebas, escriba el mensaje *I want to book a flight*. Debería aparecer la respuesta adecuada de las preguntas frecuentes.
5. Cuando termine de probar la base de conocimiento, haga clic en **&rarr; Test** para cerrar el panel de pruebas.

## Crear un bot para la base de conocimientos

La base de conocimiento es un servicio de back-end que las aplicaciones de clientes pueden usar para responder preguntas a través de una interfaz de usuario. Habitualmente, las aplicaciones de clientes son bots. Para que el bot pueda acceder a la base de conocimiento, debe publicarla como un servicio accesible desde HTTP. Puede usar Azure Bot Service para crear y hospedar un bot que utilice la base de conocimiento para responder a las preguntas de los usuarios.

1. En la parte superior de la página de QnA Maker, haga clic en **Publish**. Después, en la página **Margies Travel KB**, haga clic en **Publish**.
2. Una vez implementado el servicio, haga clic en **Create Bot**. Se abrirá Azure Portal en una nueva pestaña del navegador, para que pueda crear un bot de aplicación web en su suscripción de Azure.
3. En Azure Portal, cree un bot de aplicación web con la siguiente configuración (la mayoría se rellenarán automáticamente)
  - **Identificador de bot**: *un nombre exclusivo para el bot*
  - **Suscripción**: *su suscripción de Azure*
  - **Grupo de recursos**: *el grupo de recursos que contiene su recurso de QnA Maker*
  - **Ubicación**: *la misma ubicación de su servicio de QnA Maker*.
  - **Plan de tarifa**: F0
  - **Nombre de la aplicación**: *el mismo que **Identificador de bot** con *.azurewebsites.net* agregado automáticamente
  - **Idioma del SDK**: *seleccione C# o Node.js*
  - **Clave de autenticación de QnA**: *esta debe ser automáticamente la clave de su base de conocimiento de QnA*
  - **Ubicación/plan de servicio de aplicaciones**: *debe configurarse automáticamente con un plan y ubicación adecuados*
  - **Application Insights**: apagado
  - **App ID y contraseña de la aplicación de Microsoft**: creación automática del App ID y contraseña de la aplicación.
4. Espere hasta que se cree su bot (el icono de campana de la esquina superior derecha se animará mientras espera) Cuando aparezca el mensaje que indica que se ha completado la implementación, haga clic en **Go to resource** (o, en la página principal, haga clic en **Resource groups*, abra el grupo de recursos en el que haya creado el bot de aplicación web y haga clic sobre él).
5. En la hoja de su bot, vaya a la página **Test in Web Chat** y espere hasta que el bot muestre el mensaje **Hello and welcome!** (Puede tardar unos segundos en iniciarse).
6. Use la interfaz de pruebas de chat para comprobar que el bot responde a las preguntas de su base de conocimiento de forma correcta. Por ejemplo, escriba *I need to cancel my hotel*.
